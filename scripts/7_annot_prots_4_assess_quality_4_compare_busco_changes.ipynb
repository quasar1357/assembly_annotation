```python
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
assemblies = ['Sha', 'Ler_3', 'Cvi_0', 'C24', 'An_1']
c_buscos_assembly = [4539, 4543, 4558, 4543, 4536]  # Replace with your actual assembly scores
total_buscos = 4596
annotation_scores = [4190, 4112, 3977, 3919, 4230]  # Replace with your actual annotation scores

# Calculate relative change
relative_change = [(anno - assembly) / assembly * 100 for assembly, anno in zip(c_buscos_assembly, annotation_scores)]

# Plotting
plt.figure(figsize=(15, 6))
bars = plt.bar(assemblies, relative_change, color=['blue', 'orange', 'green', 'red', 'purple'], alpha=0.6)

# Adding labels and title
plt.xlabel('Genome Assemblies')
plt.ylabel('Relative Change in complete BUSCOs (%)')
plt.title('Complete BUSCOs')
plt.ylim(-16, 0)  # Adjust these values based on your preference
# plt.legend()

# Add annotations to the bars
for bar, assembly, anno, rel_change in zip(bars, c_buscos_assembly, annotation_scores, relative_change):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.3, -1.5, f'Assembly:\n{assembly} ({round(assembly/total_buscos*100, 2)}%)', ha='left', color='black')
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.3, bar.get_height() + 0.5, f'{anno - assembly} ({round(rel_change, 2)}%)', ha='left', color='black')
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.3, bar.get_height() - 1.2, f'Annotation:\n{anno} ({round(anno/total_buscos*100, 2)}%)', ha='left', color='black')

# Display the plot
plt.show()
```
